## Define functions
Why do we want to define functions in this case?

```python
def distance_matrix(data):
    dist=[]
    for i in range(data.shape[0]):
        dist_row = []
        for j in range(data.shape[0]):
            single_dist = sum((data[i] - data[j]) ** 2) ** (1/2)
            dist_row.append(single_dist)
        dist.append(dist_row)  
    return dist    

def closest_points(dist_matrix):
    # CODE HERE
    min_args, min_dist = (None, 9e99)
    for id_r, row in enumerate(dist_matrix):
        row_relevant = row[:id_r]
        if len(row_relevant) == 0: 
            continue
     
        row_min = min(row_relevant)

        if row_min < min_dist: 
            min_dist = row_min
            min_args = id_r 
            for id_c, col in enumerate(row_relevant): 
                if col == min_dist: 
                    min_args = id_r, id_c
    
    return min_args, min_dist
```

## Finding the `n` shortest distances

```python
dist_matrix = distance_matrix(data)
n_distances = 10

distance = []

for n in range(n_distances):
    results = closest_points(dist_matrix)
    dist_matrix[results[0][0]][results[0][1]]=9e99
    distance.append(results)

        
```
